```python
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline

DATA_PATH = os.path.join(os.path.curdir, 'names')
start = 1900
stop = 2000
name1 = 'Ruth'
name2 = 'Robert'
```

```python
name1_count = []
name2_count = []
name_years = []
# name2_count = []
for year in range(start, stop+1):
    name_years.append(year)
    file_path = os.path.join(
        DATA_PATH, f'yob{year}.txt'
    )
    file_data = pd.read_csv(
        file_path,
        names=['Name', 'Gender', 'Count'],
    )
    fn = file_data.query('Name == @name1')
    name1_count.append(fn['Count'].sum())
    fn = file_data.query('Name == @name2')
    name2_count.append(fn['Count'].sum())

plt.xlabel('Years')
plt.ylabel('Count')
plt.title(f'Name {name1}')
plt.plot(name_years, name1_count)
plt.show()
plt.title(f'Name {name2}')
plt.plot(name_years, name2_count)
```
