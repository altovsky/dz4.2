```python
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline

DATA_PATH = os.path.join(os.path.curdir, 'names')
start = 1900
stop = 2000
name1 = 'Ruth'
name2 = 'Robert'
```

```python
def plot_name_hist(name, start_year, stop_year, step):
    name_count = []
    year_list = []
    for year in range(start_year, stop_year+1, step):
        year_list.append(year)
        file_path = os.path.join(
            DATA_PATH, f'yob{year}.txt'
        )
        file_data = pd.read_csv(
            file_path,
            names=['Name', 'Gender', 'Count'],
        )
        fn = file_data.query('Name == @name')
        name_count.append(fn['Count'].sum())

    plt.xlabel('Years')
    plt.ylabel('Count')
    plt.title(f'Name {name}')
    plt.plot(year_list, name_count)
    plt.show()
```
